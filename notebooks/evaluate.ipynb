```python
from pathlib import Path
import os

if Path.cwd().name == "notebooks":
    os.chdir("..")

Path.cwd()
```

```python
import yaml

cfg = yaml.safe_load(Path("configs/eval_default.yaml").read_text())
cfg["report"].update({"out_dir": "reports/notebook_eval", "plots": True})
cfg["tracking"].update({"enabled": True, "out_dir": "experiments", "tags": ["notebook", "eval"]})

demo_config = Path("artifacts/notebooks/eval_demo.yaml")
demo_config.parent.mkdir(parents=True, exist_ok=True)
demo_config.write_text(yaml.safe_dump(cfg, sort_keys=False))
demo_config
```

```python
from dork.pipelines import run_eval

report = run_eval(str(demo_config))
[(row["suite"], row["metric"], row["value"]) for row in report["summary"]]
```
