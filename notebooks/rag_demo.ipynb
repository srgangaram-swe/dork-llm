# RAG Demo

Ingests the bundled sample docs, retrieves relevant chunks with deterministic hash embeddings, and returns a cited answer with the mock generator.

```python
from pathlib import Path
import os

if Path.cwd().name == "notebooks":
    os.chdir("..")

Path.cwd()
```

```python
import yaml

cfg = yaml.safe_load(Path("configs/rag_default.yaml").read_text())
cfg["embeddings"].update({"backend": "hash", "dim": 256})
cfg["vector_store"].update({"backend": "memory"})
cfg["generation"].update({"provider": "mock", "temperature": 0.0})
cfg["retrieval"].update({"top_k": 5, "rerank": True, "rerank_top_n": 3})

demo_config = Path("artifacts/notebooks/rag_demo.yaml")
demo_config.parent.mkdir(parents=True, exist_ok=True)
demo_config.write_text(yaml.safe_dump(cfg, sort_keys=False))
demo_config
```

```python
from dork.pipelines import ingest, query_rag

ingest_result = ingest(str(demo_config), source="data/sample_docs")
ingest_result
```

```python
answer = query_rag(str(demo_config), "What does causal masking prevent?")
answer["answer"], answer["citations"]
```
