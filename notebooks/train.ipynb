# Train Tiny GPT

A lightweight, CPU-friendly training walkthrough. It writes demo artifacts under `artifacts/notebooks/` and local tracking runs under `experiments/`.

```python
from pathlib import Path
import os

if Path.cwd().name == "notebooks":
    os.chdir("..")

Path.cwd()
```

```python
import yaml

base = yaml.safe_load(Path("configs/train_tiny_gpt.yaml").read_text())
base["tokenizer"].update({"type": "char", "vocab_size": 128, "path": "tokenizers/notebook_char.json"})
base["model"].update({"block_size": 64, "n_layer": 2, "n_head": 2, "n_embd": 64, "dropout": 0.0})
base["training"].update(
    {
        "batch_size": 8,
        "max_steps": 20,
        "eval_interval": 10,
        "eval_iters": 2,
        "learning_rate": 3.0e-3,
        "warmup_steps": 2,
        "device": "cpu",
        "dtype": "float32",
        "out_dir": "artifacts/notebooks/tiny_gpt",
    }
)
base["generation"].update({"max_new_tokens": 48, "temperature": 0.8})
base["tracking"].update({"enabled": True, "out_dir": "experiments", "tags": ["notebook", "train"]})

demo_config = Path("artifacts/notebooks/train_demo.yaml")
demo_config.parent.mkdir(parents=True, exist_ok=True)
demo_config.write_text(yaml.safe_dump(base, sort_keys=False))
demo_config
```

```python
from dork.pipelines import train_model

train_result = train_model(str(demo_config))
train_result
```

```python
from dork.pipelines import generate

print(generate(str(demo_config), "Once upon a time", max_new_tokens=48))
```
